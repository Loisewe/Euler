# EULER 17

```python
def numtowordss(num):
    string_num=str(num)
    num2words1 = {1: 'One', 2: 'Two', 3: 'Three', 4: 'Four', 5: 'Five',
                  6: 'Six', 7: 'Seven', 8: 'Eight', 9: 'Nine', 10: 'Ten',
                  11: 'Eleven', 12: 'Twelve', 13: 'Thirteen', 14: 'Fourteen',
                  15: 'Fifteen', 16: 'Sixteen', 17: 'Seventeen', 18: 'Eighteen', 19: 'Nineteen'}
    num2words2 = ['Twenty', 'Thirty', 'Forty', 'Fifty', 'Sixty', 'Seventy', 'Eighty', 'Ninety']

    if 0 < num <=19:
        return num2words1[num]
    elif 19 < num <100:
        if num%10==0:
            return num2words2[int(string_num[0])-2]
        else:
            return str(num2words2[int(string_num[0])-2] + ' ' +num2words1[int(string_num[1])])
    elif 100<= num <1000:
        if num%100==0:
            return str(num2words1[int(string_num[0])] + ' hundred')
        elif int(string_num[1:])<20:
            return str(num2words1[int(string_num[0])] + ' hundred and ' + num2words1[int(string_num[1]+string_num[2])])
        elif int(string_num[1:])%10==0:
            return str(num2words1[int(string_num[0])] + ' hundred and ' + num2words2[int(string_num[1])-2])
        elif int(string_num[1:])>20:
            return str(num2words1[int(string_num[0])] + ' hundred and ' + num2words2[int(string_num[1])-2] + ' ' +num2words1[int(string_num[2])])
    elif num==1000:
        return "One thousand" 

    
sumlett=0
for i in range (1,1001):
    for num in numtowordss(i).split(' '):
        sumlett+=len(num)
sumlett
```

# Euler 21

```python
amics=set()
def isamic(a,b):
    if a!=b:
        nums=[]
        for i in range(1,b):
            if b%i==0:
                nums.append(i)
        if sum(nums)==a: 
            amics.add(a)
            amics.add(b)
def main(n):
    for i in range(1,n+1):
        divis=[]
        if i not in amics:
            [divis.append(j) for j in range(1,i) if i%j==0]
            isamic(i,sum(divis))
            
main(10000)
            
print(sum(amics))
  
```
